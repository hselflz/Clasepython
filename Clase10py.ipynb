```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,10)
y1 = x**2
#plt.plot(x,y1,'y:',marker='*',ms=40,mec='r')#'--' Se usa para cambiar las lineas MarkerEdgeCountor
plt.xlabel('A')
plt.ylabel('B')
font1 = {'family':"serif","color":"Blue","size":15}
plt.title("Grafica de la funcion x**2", fontdict = font1)
plt.plot(x,y1,'b', linewidth='1') #MarkerFillContour
plt.grid(color = "r",linestyle = "-",linewidth = 1)
plt.show()
```

```python
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.xlim(0,10) #Lim sirve para acotar el grafico en distintas cantidas, ejemplo, de 0 a 10
plt.ylim(0,200) # y aqui de 0 a 200

x = np.linspace(0,10,10)
y1 = x**2
#plt.plot(x,y1,'y:',marker='*',ms=40,mec='r')#'--' Se usa para cambiar las lineas MarkerEdgeCountor
plt.plot(x,y1,'y:',marker='o',ms=10,mfc='b') #MarkerFillContour

plt.legend(['Función x**2'],loc = 'upper right', fontsize=10)

plt.savefig('Función de x**2')

plt.show()


```
